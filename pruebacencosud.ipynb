```python
import pandas as pd
import io
from google.colab import files

# Upload the file
uploaded = files.upload()

# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]  # Assumes only one file is uploaded

# Read the Excel file using BytesIO to handle the in-memory file
df = pd.read_excel(io.BytesIO(uploaded[filename]))

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(io.BytesIO(uploaded[filename]), sheet_name=None)

# Remove the "Ejercicio" sheet from the dictionary
if "Ejercicio" in all_sheets:
    del all_sheets["Ejercicio"]

# Concatenate the remaining DataFrames into a single DataFrame
df = pd.concat(all_sheets.values(), ignore_index=True)


# Now you can work with the DataFrame 'df'
print(df.head())
```

```python
from IPython.display import display

display(df) # Displays the DataFrame as a formatted table
display(df) # Displays the DataFrame as a formatted table # This line was causing the error due to incorrect indentation
```

```python
# Convert 'Sueldo Base' and 'Bono %' columns to numeric type
df['Sueldo Base'] = pd.to_numeric(df['Sueldo Base'], errors='coerce')
df['Bono %'] = pd.to_numeric(df['Bono %'], errors='coerce')

# Now perform the calculation
df['Bono'] = df['Sueldo Base'] * df['Bono %']
display(df)
```

```python
# Calculate 'Compensación Total' column
df['Compensación Total'] = df['Sueldo Base'] + df['Bono']

display(df)
```

```python
# Save the DataFrame to an Excel file
df.to_excel('Análisis 2024.xlsx', index=False)

# Display the DataFrame (optional)
display(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
from openpyxl.drawing.image import Image

# ... (código anterior para calcular la compensación total) ...

# Calculate average compensation
avg_compensation = df['Compensación Total'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Compensación Total Promedio'], [avg_compensation])
plt.title('Compensación Total Promedio')
plt.ylabel('Valor')
plt.savefig('compensation_chart.png')  # Save the chart as an image

# Load the Excel workbook
workbook = openpyxl.load_workbook('Análisis 2024.xlsx')

# Create a new sheet for the report
report_sheet = workbook.create_sheet(title='Reporte')

# Add the average compensation value to the report sheet
report_sheet['A1'] = 'Compensación Total Promedio'
report_sheet['B1'] = avg_compensation

# Insert the chart image
img = Image('compensation_chart.png')
report_sheet.add_image(img, 'A3')

# Save the workbook
workbook.save('Análisis 2024.xlsx')
```

```python
# Define bins and labels for 'Antigüedad'
bins = [1, 3, 5, 10, 25]
labels = ['1-3', '4-5', '6-10', '11-25']

# Create a new column with the bins
df['Tramos Antigüedad'] = pd.cut(df['Antigüedad'], bins=bins, labels=labels, include_lowest=True)

# Count unique 'Nombre' within each 'Tramos Antigüedad'
antiguedad_counts = df.groupby('Tramos Antigüedad')['Nombre'].nunique()

# Print the counts
print(antiguedad_counts)

# Create a bar chart for the counts
plt.figure(figsize=(8, 6))
antiguedad_counts.sort_index().plot(kind='bar', edgecolor='black')
plt.title('Cantidad de Personas por Tramos de Antigüedad')
plt.xlabel('Tramos de Antigüedad')
plt.ylabel('Cantidad de Personas')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.savefig('antiguedad_tramos.png')

# ... (código para cargar el libro de Excel y crear la hoja de reporte) ...

# Insert the bar chart image
img_tramos = Image('antiguedad_tramos.png')
report_sheet.add_image(img_tramos, 'A10')  # Adjust position as needed
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
from openpyxl.drawing.image import Image

# ... (código anterior para calcular la compensación total, estadísticas y tramos de antigüedad) ...

# Calculate average compensation per 'Tramos Antigüedad'
avg_compensation_per_tramo = df.groupby('Tramos Antigüedad')['Compensación Total'].mean()

# Print the average compensation per tramo
print(avg_compensation_per_tramo)

# Create a bar chart for the counts and average compensation
fig, ax1 = plt.subplots(figsize=(8, 6))

# Bar chart for counts
antiguedad_counts.sort_index().plot(kind='bar', ax=ax1, color='skyblue', edgecolor='black', label='Cantidad de Personas')
ax1.set_xlabel('Tramos de Antigüedad')
ax1.set_ylabel('Cantidad de Personas', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0)  # Keep x-axis labels horizontal

# Create a second y-axis for average compensation
ax2 = ax1.twinx()
avg_compensation_per_tramo.sort_index().plot(kind='line', ax=ax2, color='orange', marker='o', label='Compensación Promedio')
ax2.set_ylabel('Compensación Promedio', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Cantidad de Personas y Compensación Promedio por Tramos de Antigüedad')
plt.tight_layout()
plt.savefig('antiguedad_tramos_compensacion.png')

# ... (código para cargar el libro de Excel y crear la hoja de reporte) ...

# Insert the bar chart image
img_tramos_compensacion = Image('antiguedad_tramos_compensacion.png')
report_sheet.add_image(img_tramos_compensacion, 'A10')  # Adjust position as needed
```
