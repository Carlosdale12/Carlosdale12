```python
#cargar librerías necesarias para el funcionamiento del código
import pandas as pd
import io
from google.colab import files

# Upload the file
uploaded = files.upload()
```

```python
# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]  # Assumes only one file is uploaded
```

```python
# Read the Excel file using BytesIO to handle the in-memory file
df = pd.read_excel(io.BytesIO(uploaded[filename]))

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(io.BytesIO(uploaded[filename]), sheet_name=None)
```

```python
# Remove the "Ejercicio" sheet from the dictionary
if "Ejercicio" in all_sheets:
    del all_sheets["Ejercicio"]
```

```python
for sheet_name, df_sheet in all_sheets.items():
    # Verificar si hay datos faltantes en la hoja actual
    missing_data = df_sheet.isnull().sum()

    # Imprimir información sobre los datos faltantes
    print(f"Hoja: {sheet_name}")
    print(missing_data)
    print("\n")
```

```python
for sheet_name, df_sheet in all_sheets.items():
    # Verificar si la columna "Nombre" existe en la hoja
    if "Nombre" in df_sheet.columns:
        # Buscar duplicados en la columna "Nombre"
        duplicates = df_sheet["Nombre"].duplicated().sum()

        # Imprimir información sobre los duplicados
        print(f"Hoja: {sheet_name}")
        print(f"Duplicados en la columna 'Nombre': {duplicates}")
        print("\n")
    else:
        print(f"Hoja: {sheet_name}")
        print("La columna 'Nombre' no existe en esta hoja.")
        print("\n")
```

```python
for sheet_name, df_sheet in all_sheets.items():
    if "Nombre" in df_sheet.columns:
        # Obtener los nombres repetidos
        duplicate_names = df_sheet[df_sheet["Nombre"].duplicated(keep=False)]["Nombre"].unique()

        # Imprimir información sobre los duplicados
        print(f"Hoja: {sheet_name}")
        if len(duplicate_names) > 0:
            print(f"Nombres repetidos: {', '.join(duplicate_names)}")
        else:
            print("No se encontraron nombres repetidos.")
        print("\n")
    else:
        print(f"Hoja: {sheet_name}")
        print("La columna 'Nombre' no existe en esta hoja.")
        print("\n")
```

```python
names_to_check = ["Luis Torres", "Paula Fernández", "Mariana Mendoza"]

for sheet_name, df_sheet in all_sheets.items():
    if "Nombre" in df_sheet.columns:
        for name in names_to_check:  # Este bloque debe estar indentado
            # Filtrar filas con el nombre actual
            filtered_rows = df_sheet[df_sheet["Nombre"] == name]
            # Verificar si hay más de una fila con ese nombre
            if filtered_rows.shape[0] > 1:
                # Verificar si todas las filas son iguales (excepto la columna "Nombre")
                if filtered_rows.drop(columns=["Nombre"]).eq(filtered_rows.drop(columns=["Nombre"]).iloc[0]).all(axis=None):
                    print(f"Hoja: {sheet_name}")
                    print(f"Nombre: {name} - Datos de otras columnas iguales")
                    print("\n")
                else:
                    print(f"Hoja: {sheet_name}")
                    print(f"Nombre: {name} - Datos de otras columnas diferentes")
                    print("\n")
    else: # Este bloque también debe estar indentado
        print(f"Hoja: {sheet_name}")
        print("La columna 'Nombre' no existe en esta hoja.")
        print("\n")
```

```python
for sheet_name, df_sheet in all_sheets.items():
    # Verificar si hay datos faltantes en la hoja actual
    missing_data = df_sheet.isnull().sum()

    # Imprimir información sobre los datos faltantes
    print(f"Hoja: {sheet_name}")
    print(missing_data)
    print("\n")
```

```python
# Suponiendo que 'all_sheets' es tu diccionario de DataFrames

sheet_names = list(all_sheets.keys())  # Obtener los nombres de las hojas

for i in range(1, len(sheet_names)):
    current_sheet = sheet_names[i]
    previous_sheet = sheet_names[i - 1]

    # Iterar sobre las columnas del DataFrame actual
    for column in all_sheets[current_sheet].columns:
        # Buscar valores faltantes en la columna actual
        missing_indices = all_sheets[current_sheet][column][all_sheets[current_sheet][column].isnull()].index

        # Rellenar los valores faltantes con el valor de la hoja anterior + 0.1
        for index in missing_indices:
            try:
                all_sheets[current_sheet].loc[index, column] = all_sheets[previous_sheet].loc[index, column] + 0.1
            except KeyError:
                # Manejar el caso donde la fila o columna no existe en la hoja anterior
                print(f"Advertencia: Fila {index} o columna '{column}' no encontrada en la hoja anterior '{previous_sheet}'")
                # Puedes elegir otro valor para rellenar en este caso, por ejemplo, 0
                all_sheets[current_sheet].loc[index, column] = 0
```

```python
all_data = pd.DataFrame()  # Crear un DataFrame vacío para almacenar todos los datos

for sheet_name, df_sheet in all_sheets.items():
    df_sheet['sheet_name'] = sheet_name  # Agregar una columna con el nombre de la hoja
    all_data = pd.concat([all_data, df_sheet], ignore_index=True)

print("DataFrame combinado creado.")
# Ahora puedes acceder al DataFrame combinado con la variable 'all_data'
all_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir los límites de los tramos de antigüedad
bins = [0, 1, 3, 5, 10, float('inf')]  # Límites de los tramos en años
labels = ['0-1 año', '1-3 años', '3-5 años', '5-10 años', 'Más de 10 años']  # Etiquetas para los tramos

# Crear la columna 'Tramo de Antigüedad'
all_data['Tramo de Antigüedad'] = pd.cut(all_data['Antigüedad'], bins=bins, labels=labels, right=False)

# Calcular la compensación promedio por tramo de antigüedad
compensacion_promedio_por_tramo = all_data.groupby('Tramo de Antigüedad')['Compensación Total'].mean()

# Crear el gráfico
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura
compensacion_promedio_por_tramo.plot(kind='bar', color='skyblue')
plt.xlabel('Tramo de Antigüedad')
plt.ylabel('Compensación Total Promedio')
plt.title('Compensación Total Promedio por Tramo de Antigüedad')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
!pip install reportlab

import matplotlib.pyplot as plt
import pandas as pd
from io import BytesIO
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet

# Asumiendo que tienes un DataFrame llamado 'all_data' con los datos
# ... (código para generar los datos y el DataFrame 'all_data') ...


# ----> Crea la columna 'Mes' si no existe <----
if 'Mes' not in all_data.columns:
    all_data['Fecha de Ingreso'] = pd.to_datetime(all_data['Fecha de Ingreso'])  # Convertir a datetime
    all_data['Mes'] = all_data['Fecha de Ingreso'].dt.month_name()  # Extraer el nombre del mes
# ----> Fin de la creación de la columna 'Mes' <----


# Define los colores institucionales de Cencosud (reemplázalos con los colores reales)
color_principal = "#003399"  # Azul Cencosud (ejemplo)
color_secundario = "#FFCC00"  # Amarillo Cencosud (ejemplo)


# ----> Genera el gráfico con los colores institucionales <----
compensacion_por_mes = all_data.groupby('Mes')['Compensación Total'].sum()
ranking_meses = compensacion_por_mes.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(ranking_meses.index, ranking_meses.values, color=color_principal)
plt.xlabel('Mes')
plt.ylabel('Compensación Total')
plt.title('Ranking de Meses con Mayores Compensaciones', color=color_secundario)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# ---->  Ranking de mayor antigüedad por persona <----
ranking_antiguedad = all_data.groupby('Nombre')['Antigüedad'].max().sort_values(ascending=False)

plt.figure(figsize=(12, 6))  # Ajustar el tamaño de la figura
plt.barh(ranking_antiguedad.index, ranking_antiguedad.values, color=color_principal)
plt.xlabel('Antigüedad (años)')
plt.ylabel('Nombre')
plt.title('Ranking de Mayor Antigüedad por Persona', color=color_secundario)
plt.gca().invert_yaxis()  # Invertir el eje y para que el ranking sea descendente
plt.tight_layout()  # Ajustar el diseño para evitar que las etiquetas se superpongan

# ----> Fin de la generación del gráfico <----


# ----> Integra el gráfico en el informe PDF <----
doc = SimpleDocTemplate("reporte_cencosud.pdf", pagesize=letter)
styles = getSampleStyleSheet()
story = []
story.append(Paragraph("Informe de Análisis de Datos", styles['h1']))
story.append(Spacer(1, 12))

# Guarda el gráfico en memoria
imgdata = BytesIO()
plt.savefig(imgdata, format='png')
imgdata.seek(0)

# Agrega el gráfico al informe
story.append(Image(imgdata, width=400, height=300))
story.append(Spacer(1, 12))

# ----> Agrega resultados de ejemplo <----
# 1. Ranking de meses con mayores compensaciones
story.append(Paragraph("Ranking de meses con mayores compensaciones:", styles['h2']))
for i, (mes, valor) in enumerate(ranking_meses.items()):
    story.append(Paragraph(f"{i+1}. {mes}: {valor}", styles['Normal']))
story.append(Spacer(1, 12))

# 2. Ranking de mayor antigüedad por persona
story.append(Paragraph("Ranking de mayor antigüedad por persona:", styles['h2']))
for i, (nombre, antiguedad) in enumerate(ranking_antiguedad.items()):
    story.append(Paragraph(f"{i+1}. {nombre}: {antiguedad} años", styles['Normal']))
story.append(Spacer(1, 12))
# ----> Fin de resultados de ejemplo <----

# ... (agregar otros elementos al informe) ...

doc.build(story)
print("Informe generado: reporte_cencosud.pdf")
# ----> Fin de la integración del gráfico <----
```
