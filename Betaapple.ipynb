```python
# -*- coding: utf-8 -*-
"""Trabajo_2_Finanzas_3_CALBULLANCA_YASMIN_A04

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1oaLzGfRYNYBpJEurFwtZhMc08Jx2tUPv
"""

!pip install --upgrade yfinance --quiet

import yfinance as yf
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


# Definir el símbolo de la acción y el período de tiempo
symbol_empresa = "AAPL"
symbol_mercado = "^NDX"
start_date = "2020-03-28"
end_date = "2025-03-28"

# Descargar datos desde Yahoo Finance
df_empresa = yf.download(symbol_empresa, start=start_date, end=end_date)
df_mercado = yf.download(symbol_mercado, start=start_date, end=end_date)


# Mostrar las primeras filas del DataFrame
print(df_empresa.head())
print(df_mercado.head())



## Descarga de Datos desde Yahoo Finance

# Datos empresa
df_empresa_d = yf.download(symbol_empresa, start=start_date, interval="1d")
df_empresa_w = yf.download(symbol_empresa, start=start_date, interval="1wk")
df_empresa_m = yf.download(symbol_empresa, start=start_date, interval="1mo")

# Datos mercado
df_mercado_d = yf.download(symbol_mercado, start=start_date, interval="1d")
df_mercado_w = yf.download(symbol_mercado, start=start_date, interval="1wk")
df_mercado_m = yf.download(symbol_mercado, start=start_date, interval="1mo")

## Calculos ##
# Función para calcular el beta
def calcular_beta(df_empresa, df_mercado):
    df_empresa['Return'] = df_empresa['Close'].pct_change()
    df_mercado['Return'] = df_mercado['Close'].pct_change()

    df_empresa = df_empresa.dropna()
    df_mercado = df_mercado.dropna()

    slope, _, _, _, _ = stats.linregress(df_mercado['Return'], df_empresa['Return'])
    return slope

# Calcular Beta para diferentes frecuencias
beta_daily = calcular_beta(df_empresa_d, df_mercado_d)
beta_weekly = calcular_beta(df_empresa_w, df_mercado_w)
beta_monthly = calcular_beta(df_empresa_m, df_mercado_m)

# Mostrar los resultados
print(f"Beta diario: {beta_daily}")
print(f"Beta semanal: {beta_weekly}")
print(f"Beta mensual: {beta_monthly}")

## Resumen en tabla ##

# Crear un DataFrame con los valores de Beta
summary_data = {'Frecuencia': ['Diaria', 'Semanal', 'Mensual'],
    'Beta': [beta_daily, beta_weekly, beta_monthly]}
summary_df = pd.DataFrame(summary_data)

# Mostrar la tabla resumen
print(summary_df)
```
