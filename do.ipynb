```python
import matplotlib.pyplot as plt

dimensiones = [
    'Liderazgo', 'Comunicacin interna', 'Ambiente y relaciones',
    'Desarrollo y capacitacin', 'Bienestar y conciliacin', 'Identidad'
]
puntajes = [3.6, 3.2, 3.9, 3.4, 3.1, 4.2]

plt.figure(figsize=(10, 6))
plt.barh(dimensiones, puntajes, color='mediumpurple')
plt.title('Resultados Encuesta Clima Laboral CENABAST')
plt.xlabel('Puntaje promedio (1 a 5)')
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
