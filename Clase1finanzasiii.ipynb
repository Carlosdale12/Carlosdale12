```python
# Instalar librerías si es necesario
!pip install yfinance openpyxl

# Importar librerías
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definir el símbolo y el rango de fechas
symbol = "AAPL"
start_date = "2020-01-01"

# Descargar los datos
df = yf.download(symbol, start=start_date)

# Mostrar las primeras filas del dataframe
df.head()
```

```python
print(df.columns)
```

```python
# Accessing the 'Close' data for AAPL:
close_prices = df['Close', 'AAPL']

# Or, using .loc:
close_prices = df.loc[:, ('Close', 'AAPL')]

# Calculate the daily returns:
df["Rentabilidad"] = close_prices.pct_change()
```

```python
display(df)
```

```python
df['Close'].plot()

plt.title('AAPL Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
